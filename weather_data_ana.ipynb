### Rename the column name Weather to "Weather Condition"

### What is mean visibility

```python
data["Visibility_km"].mean()
```

### What is the variance of "Relative Humidity" in the data?

### Here we have found the exact match of "Snow" but some values contain "Snow" as partial match. How to find those instances? 

```python
data.groupby(by = "Weather Condition").get_group("Snow")
```

```python
data["Weather Condition"].value_counts("Snow")
```

### Find all the instances when wind speed is above 24 and visiblility is 25.

```python
data[(data["Wind Speed_km/h"]>24) & (data["Visibility_km"] == 25)]
```

### Find all the instances where weather is clear or visibility is above 40.

```python
data[(data["Weather Condition"]=="Clear") | (data["Visibility_km"]>40)]
```

### Find all the instances when weather is clear and relative humidity is greater than 50 or visibility is above 40

```python
data[(data["Weather Condition"]=="Clear") & (data["Rel Hum_%"]>50) |(data["Visibility_km"] > 40)]
```
